# Stack plot

```python
import matplotlib.pyplot as plt
days = [1,2,3,4,5]
sleep = [4,6,8,9,10]
eating = [6,4,8,2,2]
working = [6,8,4,5,4]
playing = [7,4,1,4,3]
plt.stackplot(days, sleep, eating, working, playing, colors = ['c','r','b','g'])
plt.xlabel("Days in a Week")
plt.ylabel("Hours of Activities")
plt.title("!! Activities in a Week !!")
plt.show()
```
